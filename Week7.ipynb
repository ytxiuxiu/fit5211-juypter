```python
class BST:
    def __init__(self, val):
        self.l_child = None
        self.r_child = None
        self.data = val
        
    def getRightChild(self):
        return self.r_child

    def getLeftChild(self):
        return self.l_child
    
    def getNodeValue(self):
        return self.data

# External Methods
def insertNode(root, node):
    if root is None:
        root = node
    else:
        if root.data > node.data:
            if root.l_child is None:
                root.l_child = node
            else:
                insertNode(root.l_child, node)
        else:
            if root.r_child is None:
                root.r_child = node
            else:
                insertNode(root.r_child, node)

def in_order_print(root):
    if not root:
        return
    in_order_print(root.l_child)
    print(root.data)
    in_order_print(root.r_child)

def pre_order_print(root):
    if not root:
        return        
    print(root.data)
    pre_order_print(root.l_child)
    pre_order_print(root.r_child)   
    
def post_order_print(root):
    if not root:
        return        
    pre_order_print(root.l_child)
    pre_order_print(root.r_child)  
    print(root.data)
        
def findValue(tree, v):
    if not tree:
        return False
    
    if tree.getNodeValue() == v:
        return True
    
    left = findValue(tree.getLeftChild(), v)
    right = findValue(tree.getRightChild(), v)
    return left or right

def findMinimum(tree):
    if not tree:
        return None
    if not tree.getLeftChild():
        return tree.getNodeValue()
    
    return findMinimum(tree.getLeftChild())

def findMaximum(tree):
    if not tree:
        return None
    if not tree.getRightChild():
        return tree.getNodeValue()
    
    return findMaximum(tree.getRightChild())

def sumNodes(tree):
    if not tree:
        return 0
    
    left = findSum(tree.getLeftChild())
    right = findSum(tree.getRightChild())
    return left + right + tree.getNodeValue()

def countNodes(tree):
    if not tree:
        return 0
    
    left = countNodes(tree.getLeftChild())
    right = countNodes(tree.getRightChild())
    return left + right + 1

def measureHeight(tree):
    if not tree:
        return -1
    
    left = measureHeight(tree.getLeftChild())
    right = measureHeight(tree.getRightChild())
    return max(left, right) + 1

# Driver Program
r = BST(3)
insertNode(r, BST(7))
insertNode(r, BST(1))
insertNode(r, BST(5))

print("in order:")
in_order_print(r)

print("find value:")
print(findValue(r, 1))
print(findValue(r, 5))
print(findValue(r, 7))
print(findValue(r, 6))

print("find sum:")
print(findSum(r))

print("find min:")
print(findMinimum(r))

print("find max:")
print(findMaximum(r))

print("count nodes:")
print(countNodes(r))

print("measure height:")
print(measureHeight(r))
```
