```python
class BinaryTree:
    def __init__(self, key):
        self.key = key
        self.leftChild = None
        self.rightChild = None
        
    def insertLeft(self, tree):
        assert self.leftChild == None
        self.leftChild = tree
        
    def insertRight(self, tree):
        assert self.rightChild == None
        self.rightChild = tree
        
    def getRightChild(self):
        return self.rightChild

    def getLeftChild(self):
        return self.leftChild
    
    def getNodeValue(self):
        return self.key
    
    def preorderPrint(self):
        print(self.getNodeValue(), end = ',')
        if self.getLeftChild():
            self.getLeftChild().preorderPrint()
        if self.getRightChild():
            self.getRightChild().preorderPrint()

    def inorderPrint(self):
        if self.getLeftChild():
            self.getLeftChild().inorderPrint()
        print(self.getNodeValue(), end = ',')
        if self.getRightChild():
            self.getRightChild().inorderPrint()
            
    def postorderPrint(self):
        if self.getLeftChild():
            self.getLeftChild().postorderPrint()
        if self.getRightChild():
            self.getRightChild().postorderPrint()
        print(self.getNodeValue(), end = ',')
        
def measureHeight(tree):
    if not tree:
        return -1
    
    left = measureHeight(tree.getLeftChild())
    right = measureHeight(tree.getRightChild())
    return max(left, right) + 1

def findSum(tree):
    if not tree:
        return 0
    
    left = findSum(tree.getLeftChild())
    right = findSum(tree.getRightChild())
    return left + right + tree.getNodeValue()
        
def findValue(tree, v):
    if not tree:
        return False
    
    if tree.getNodeValue() == v:
        return True
    
    left = findValue(tree.getLeftChild(), v)
    right = findValue(tree.getRightChild(), v)
    return left or right
    
def findValueAtMinimumDepth(tree, v):
    if not tree:
        return -1
    if not findValue(tree, v):
        return None
    
    if tree.getNodeValue() == v:
        return 0
    else:
        left = findValueAtMinimumDepth(tree.getLeftChild(), v)
        left = 0 if not left else left
        right = findValueAtMinimumDepth(tree.getRightChild(), v)
        right = 0 if not right else right
        return min(left, right) + 1
    
n1 = BinaryTree(5)
n2 = BinaryTree(10)
n3 = BinaryTree(12)
n3.insertLeft(n1)
n3.insertRight(n2)
n4 = BinaryTree(9)
n5 = BinaryTree(8)
n5.insertLeft(n3)
n5.insertRight(n4)
n6 = BinaryTree(5)
n7 = BinaryTree(6)
n8 = BinaryTree(7)
n8.insertLeft(n6)
n8.insertRight(n7)
n9 = BinaryTree(7)
n9.insertLeft(n5)
n9.insertRight(n8)
t1 = n9

print('tree 1')
t1.preorderPrint()
print()
t1.inorderPrint()
print()
t1.postorderPrint()
print()

print(measureHeight(t1))

print(findSum(t1)) #69

print('find value 1: {}'.format(findValue(t1, 1)))
print('find value 7: {}'.format(findValue(t1, 7)))
print('find value 8: {}'.format(findValue(t1, 8)))
print('find value 5: {}'.format(findValue(t1, 5)))
print('find value 9: {}'.format(findValue(t1, 9)))
print('find value 12: {}'.format(findValue(t1, 12)))
print('find value 6: {}'.format(findValue(t1, 6)))

print('find 7: {}'.format(findValueAtMinimumDepth(t1, 7)))
print('find 5: {}'.format(findValueAtMinimumDepth(t1, 5)))
print('find 2: {}'.format(findValueAtMinimumDepth(t1, 2)))

n1 = BinaryTree(5)
n2 = BinaryTree(10)
n3 = BinaryTree(12)
n3.insertLeft(n1)
n3.insertRight(n2)
n4 = BinaryTree(8)
n4.insertLeft(n3)
n5 = BinaryTree(5)
n6 = BinaryTree(5)
n6.insertLeft(n5)
n7 = BinaryTree(10)
n8 = BinaryTree(6)
n8.insertLeft(n7)
n9 = BinaryTree(7)
n9.insertLeft(n6)
n9.insertRight(n8)
n10 = BinaryTree(7)
n10.insertLeft(n4)
n10.insertRight(n9)
t2 = n10

print('tree 2')
t2.preorderPrint()
print()
t2.inorderPrint()
print()
t2.postorderPrint()
print()
print(measureHeight(t2))
print(findSum(t2)) #75
```

```python
class BinaryTree:
    def __init__(self, rootObj):
        self.key = rootObj
        self.leftChild = None
        self.rightChild = None
        
    def insertLeft(self, newNode):
        if self.leftChild == None:
            self.leftChild = BinaryTree(newNode)
        else:
            t = BinaryTree(newNode)
            t.leftChild = self.leftChild
            self.leftChild = t

    def insertRight(self, newNode):
        if self.rightChild == None:
            self.rightChild = BinaryTree(newNode)
        else:
            t = BinaryTree(newNode)
            t.rightChild = self.rightChild
            self.rightChild = t

    def getRightChild(self):
        return self.rightChild

    def getLeftChild(self):
        return self.leftChild
    
    def getNodeValue(self):
        return self.key
    
    def preorderPrint(self):
        print(self.getNodeValue(), end = ',')
        if self.getLeftChild():
            self.getLeftChild().preorderPrint()
        if self.getRightChild():
            self.getRightChild().preorderPrint()

    def inorderPrint(self):
        if self.getLeftChild():
            self.getLeftChild().inorderPrint()
        print(self.getNodeValue(), end = ',')
        if self.getRightChild():
            self.getRightChild().inorderPrint()
            
    def postorderPrint(self):
        if self.getLeftChild():
            self.getLeftChild().postorderPrint()
        if self.getRightChild():
            self.getRightChild().postorderPrint()
        print(self.getNodeValue(), end = ',')
        
t1 = BinaryTree(7)
t1.insertLeft(8)
t1.insertRight(7)
t1.getLeftChild().insertLeft(12)
t1.getLeftChild().insertRight(9)
t1.getRightChild().insertLeft(5)
t1.getRightChild().insertLeft(6)
t1.getLeftChild().getLeftChild().insertLeft(5)
t1.getLeftChild().getLeftChild().insertRight(10)

print('tree 1')
t1.preorderPrint()
print()
t1.inorderPrint()
print()
t1.postorderPrint()
print()

t2 = BinaryTree(7)
t2.insertLeft(8)
t2.insertRight(7)
t2.getLeftChild().insertLeft(12)
t2.getRightChild().insertLeft(5)
t2.getRightChild().insertRight(6)
t2.getLeftChild().getLeftChild().insertLeft(5)
t2.getLeftChild().getLeftChild().insertRight(10)
t2.getRightChild().getLeftChild().insertLeft(5)
t2.getRightChild().getRightChild().insertLeft(10)

print('tree 2')
t2.preorderPrint()
print()
t2.inorderPrint()
print()
t2.postorderPrint()
print()
```
